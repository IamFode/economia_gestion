# Optimización de una variable

```python
# library
from sympy import Symbol
from scipy.misc import derivative
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

## 9.2. Problemas

### 1

$$y=-2.05+1.06x-0.04x^2$$

```python
# Primera derivada
x = Symbol('x') 
y = -2.5 + 1.06*x - 0.04*x**2
y.diff(x)
```

```python
# gráfica
x = np.arange(13,13.5,0.01)
fig,ax = plt.subplots()
f = -2.5 + 1.06*x - 0.04*x**2
ax.plot(x,f)
ax.set(xlabel="Total trabajo semanal (miles de horas)",ylabel="carne producida (millones de libras)")
plt.show()
```

```python
#función
func = lambda x: -2.5 + 1.06*x - 0.04*x**2
# Total de carne producida en millones de libras
func(13.25)
```

### 2

$$h(x)=\dfrac{8x}{3x^2+4}$$

```python
# Primera derivada
x = Symbol('x') 
y = (8*x)/(3*x**2+4) 
y.diff(x)
```

```python
# gráfica
x = np.arange(-1.5,1.5,0.01)
fig,ax = plt.subplots()
f = (8*x)/(3*x**2+4)
ax.plot(x,f)
plt.show()
```

```python
#función
func = lambda x: (8*x)/(3*x**2+4)
# máximo 
print(func(np.sqrt(4/3)))
#mínimo
print(func(-np.sqrt(4/3)))
```
